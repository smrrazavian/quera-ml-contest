        در این سوال شما می‌توانید از هر تکنیک پیش‌پردازش/مهندسی ویژگی دلخواهتان، استفاده کنید.
    <br>
    تکنیک‌هایی که استفاده می‌کنید به شکل مستقیم مورد ارزیابی توسط سامانه داوری قرار <b>نمی‌گیرند.</b> بلکه همه آن‌ها در دقت مدل شما تاثیر خواهند گذاشت؛ بنابراین هر چه پیش‌پردازش/مهندسی ویژگی بهتری انجام دهید تا دقت مدل بهبود پیدا کند، امتیاز بیشتری از این سوال کسب خواهید کرد.

</font>
</p>

```python
from sklearn.preprocessing import MinMaxScaler

df_train = pd.read_csv("../data/train.csv")
scaler = MinMaxScaler()


def preprocessing(df: pd.DataFrame, type: str = "train") -> pd.DataFrame:
    df.drop(
        ["StudentID", "Age", "Volunteering", "Gender"],
        axis=1,
        inplace=True,
    )
    if type == "test":
        df["ParentalEducation"].fillna(df["ParentalEducation"].mode()[0], inplace=True)
        df["ParentalSupport"].fillna(df["ParentalSupport"].mode()[0], inplace=True)

    df["Ethnicity"] = df["Ethnicity"].fillna("Unknown")

    df["ParentalEducation"] = df["ParentalEducation"].map(
        {"High School": 0, "Some College": 1, "Bachelor": 3, "Higher": 4}
    )
    df["ParentalSupport"] = df["ParentalSupport"].map(
        {"Low": 0, "Moderate": 1, "High": 2, "Very High": 3}
    )

    df_encoded = pd.get_dummies(df, columns=["Ethnicity"], prefix="Ethnicity")
    one_hot_columns = [col for col in df_encoded.columns if "Ethnicity_" in col]
    df_encoded[one_hot_columns] = df_encoded[one_hot_columns].astype(int)

    # df_encoded["Age"] = scaler.fit_transform(df_encoded[["Age"]])
    df_encoded["StudyTimeWeekly"] = scaler.fit_transform(df[["StudyTimeWeekly"]])

    return df_encoded


df_train = preprocessing(df_train)
df_train
```

```python
X = df_train.drop(columns=['GPA'])
y = df_train['GPA']
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
import pandas as pd


param_grid = {
    "n_estimators": [100, 200, 300, 1000],
    "max_depth": [3, 5, 7],
    "learning_rate": [0.01],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8],
}

xgb_model = xgb.XGBRegressor(objective="reg:squarederror")

grid_search = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    cv=2,
    scoring="r2",
    verbose=1,
    n_jobs=1,
)

grid_search.fit(X, y)

print("Best Hyperparameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)

best_model = grid_search.best_estimator_
```
