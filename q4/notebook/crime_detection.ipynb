```python
from pandas import DataFrame
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MultiLabelBinarizer
import pandas as pd
import numpy as np


label_encoder = LabelEncoder()
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()


def df_processor(df: DataFrame, df_type: str = "train") -> DataFrame:
    "Pre-process the dataframe"
    # Handling MoCodes
    df["Mocodes"] = [
        str(x).split(" ") if not pd.isnull(x) else [] for x in df["Mocodes"]
    ]
    df["Mocodes"] = [
        x if isinstance(x, list) else [x] if not pd.isnull(x) else []
        for x in df["Mocodes"]
    ]

    mlb = MultiLabelBinarizer()
    df_encoded = pd.DataFrame(
        mlb.fit_transform(df["Mocodes"]), columns=mlb.classes_, index=df.index
    )
    df_encoded = df_encoded.add_prefix("Mocode_")
    df = pd.concat([df, df_encoded], axis=1)

    df = df.drop(
        [
            "DR_NO",
            "Crm Cd 1",
            "Crm Cd 2",
            "Crm Cd 3",
            "Crm Cd 4",
            "Rpt Dist No",
            "Crm Cd Desc",
            "Premis Cd",
            "Premis Desc",
            # "Weapon Used Cd",
            "Status Desc",
            "Weapon Desc",
            "LOCATION",
            "Mocodes",
            # "AREA",
            "AREA NAME",
            "Cross Street",
            # "Date Rptd",
            # "DATE OCC",
            # "TIME OCC",
            "LAT",
            "LON",
        ],
        axis=1,
        errors="ignore",
    )
    df["Vict Age"] = df["Vict Age"].replace(0, np.nan)
    df["Vict Age"] = df["Vict Age"].fillna(df["Vict Age"].median())
    df["Vict Descent"] = df["Vict Descent"].fillna(df["Vict Descent"].mode()[0])
    df["Weapon Used Cd"] = df["Weapon Used Cd"].fillna(100)

    # Date handling
    df["DATE OCC"] = pd.to_datetime(df["DATE OCC"]).map(pd.Timestamp.timestamp)
    df["Date Rptd"] = pd.to_datetime(df["Date Rptd"]).map(pd.Timestamp.timestamp)
    df["time_diff"] = df["Date Rptd"] - df["DATE OCC"]

    norm_cols = ["Vict Age", "DATE OCC", "time_diff", "Date Rptd"] #"LAT", "LON"
    for item in norm_cols:
        df[item] = scaler.fit_transform(df[[item]])

    # One hot encoding
    # Time of day handling
    df["hour"] = df["TIME OCC"] // 100

    def classify_time_of_day(hour):
        if 5 <= hour < 12:
            return "Morning"
        elif 12 <= hour < 17:
            return "Afternoon"
        elif 17 <= hour < 21:
            return "Evening"
        else:
            return "Night"

    df["time_of_day"] = df["hour"].apply(classify_time_of_day)

    init_one_hot_columns = [
        "Vict Sex",
        "Vict Descent",
        "Status",
        "Weapon Used Cd",
        "time_of_day",
        "AREA"
    ]
    for item in init_one_hot_columns:

        df_train_encoded = pd.get_dummies(df, columns=[item], prefix=item)
        one_hot_columns = [col for col in df_train_encoded.columns if f"{item}_" in col]
        df_train_encoded[one_hot_columns] = df_train_encoded[one_hot_columns].astype(
            int
        )

        df = df.join(df_train_encoded[one_hot_columns])
        df.drop(item, axis=1, inplace=True)

    if df_type == "train":
        df["Crm Cd"] = label_encoder.fit_transform(df["Crm Cd"])
    return df


df = pd.read_csv("../data/train.csv")
df = df_processor(df)
df
```

```python
X = df.drop(columns=["Crm Cd"])
y = df["Crm Cd"]
```

```python
# Grid Search
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, f1_score
import pandas as pd

dtrain = xgb.DMatrix(X, label=y)
param_grid = {
    "n_estimators": [300],
    "max_depth": [10],
    "learning_rate": [0.1],
}

xgb_model = xgb.XGBClassifier(
    eval_metric="mlogloss", random_state=42, tree_method="hist", device="cuda"
)

grid_search = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    cv=3,
    scoring="f1_macro",
    verbose=2,
    n_jobs=2,
)

grid_search.fit(X, y)

print("Best Hyperparameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)

best_model = grid_search.best_estimator_
```

```python
# Grid Search Catboost
from sklearn.model_selection import GridSearchCV
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, f1_score
import pandas as pd

param_grid = {
    "iterations": [1000],
    "depth": [8],
    "learning_rate": [0.1,],
    "subsample": [1],
}

catboost_model = CatBoostClassifier(
    silent=True,
    random_state=32,
    bootstrap_type="Bernoulli",
    task_type="GPU",
    devices="0",
)

grid_search_cat = GridSearchCV(
    estimator=catboost_model,
    param_grid=param_grid,
    cv=3,
    scoring="f1_macro",
    verbose=2,
    n_jobs=1,
)

grid_search_cat.fit(X, y)

print("Best Hyperparameters: ", grid_search_cat.best_params_)
print("Best Score: ", grid_search_cat.best_score_)

best_model_cat = grid_search_cat.best_estimator_
```

```python
columns = list(df.columns)
columns.remove('Crm Cd')

len(columns)
```

```python
# predict test samples
test_data = pd.read_csv('../data/test.csv')
test_data = df_processor(test_data, "test")
for col in columns:
    if col not in test_data.columns:
        test_data[col] = 0
print(len(test_data.columns))
X_test_new = test_data[columns]
y_pred_new = best_model_cat.predict(X_test_new)

predicted_crm = label_encoder.inverse_transform(y_pred_new)

test_data['Crm Cd'] = predicted_crm
submission = test_data["Crm Cd"]
submission

```
